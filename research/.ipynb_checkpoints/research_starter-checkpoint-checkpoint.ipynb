```python
import pandas as pd

train_data = pd.read_csv('./data/train.csv')
test_data = pd.read_csv('./data/test.csv')

print(train_data.head())
print(test_data.head())
```

## 2) Investigate the Dataset
In the datasets, you're given columns of time and A through N, each of which represent some sort of real-life market quantity. In the train dataset, you're also given Y1 and Y2, real-life market quantities you'd like to predict in terms of time and A through N. You're not given Y1 and Y2 in the test set, because this is what you're being asked to predict.

Let's do some exploration of the relationships of A - N and Y1. In particular, let's look at the relationship between C and Y1:

```python
import seaborn as sns
plt.figure(figsize = (12, 10))
sns.heatmap(train_data.corr(), annot=True)
plt.show()

# Good Predictions for Y_1 that have decent linear correlation with it
# Feature C, E, G (highest: 0.81), H, J, M, N (lowest: 0.6)
#  G, M, J, C , E, H
# High autocorrelation: Y_2, L, K, I, F, D, B, A, 

# Mutual Information: when you know Y it tells you the value of X
# High
```

```python
#plotting Y_1 and Y_2 versus time
fig, axes = plt.subplots(1, 2, figsize = (12, 4))

axes[0].plot(train_data['time'], train_data['Y1'])
axes[1].plot(train_data['time'], train_data['Y2'])


#Y1 is centered around 0 and is looking pretty stationary
# maybe try looking at autocorrelations?
```

```python
fig, axes = plt.subplots(1, 1, figsize = (8, 4))
axes.plot(train_data['time'].iloc[1:], train_data['Y2'].diff().dropna())
```

```python
# Analyzing Autocovariance
import numpy as np
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

fig, ax = plt.subplots(1, 2, figsize = (15, 4))

temp = plot_acf(train_data['Y1'], lags = 10000, ax = ax[0])
temp = plot_acf(train_data['Y2'], lags = 20000, ax = ax[1])

bound = 1.96 / np.sqrt(len(train_data['Y1']))
ax[0].set_ylim(-0.02, 0.02)
ax[0].hlines([bound, -bound], xmin = 0, xmax = 10000, colors = 'red')
ax[1].set_ylim(0, 0.25)
ax[1].set_xlim(0, 20000)
# for k in range(0, 30):
#     print(train_data['Y1'].autocorr(lag=k))
    #print(train_data['Y1'].iloc[k:].cov(train_data['Y1'].shift(k)))

#

```

```python
from scipy.signal import find_peaks
from statsmodels.tsa.stattools import acf
nlags = 20000
acf_vals = acf(train_data['Y2'], nlags=nlags)

lags = np.arange(1, nlags + 1)
acf1 = acf_vals[1:]

peak_idx, props = find_peaks(acf1, prominence = 0.01, distance = 50)
print(peak_idx, props)
peak_lags = lags[peak_idx]

peak_diffs = np.diff(peak_lags)
season_est_1 = np.median(peak_diffs)
season_est_2 = np.mean(peak_diffs)

peak_diffs, season_est_1, season_est_2
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

s = 96
y = pd.Series(train_data['Y2'].values, index = pd.RangeIndex(len(train_data['Y2']))
model = SARIMAX(y, order = (1, 1, 1), seasonal_order = (1, 1, 1, s), enforce_stationarity = False, enforce_invertibility=False)
res = model.fit(disp = False)
print(res.summary())
```

```python
fig, ax = plt.subplots(1, 1, figsize = (15, 4))

temp = plot_acf(train_data['Y2'].diff().dropna(), lags = 20000, ax = ax)

bound = 1.96 / np.sqrt(len(train_data['Y2']))
ax.hlines([bound, -bound], xmin = 0, xmax = 20000, colors = 'red')
ax.set_ylim(-0.02, 0.02)
ax.set_xlim(0, 20000)

```

```python
# What are we going to do here?
# We will look at the differences as a new thing to predict
train_data['Y2_diff'] = train_data['Y2'].diff().dropna()
plt.figure(figsize = (12, 10))
sns.heatmap(train_data.corr(), annot=True)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(train_data['C'], train_data['Y1'])
plt.xlabel('C')
plt.ylabel('Y1')
plt.title('Relationship between C and Y1')
plt.show()
```

```python
# Calculate correlation between C and Y1
correlation = train_data['C'].corr(train_data['Y1'])
print(f"Correlation between C and Y1: {correlation:.4f}")
```

Clearly there's a strong relationship between C and Y1. You should definitely use C to predict Y1!

## 3) Submit Predictions
In order to submit predictions, we need to make a CSV file with three columns: id, Y1, and Y2. In the below example, we let our predictions of Y1 and Y2 be the means of Y1 and Y2 in the train set.

```python
preds = test_data[['id']]
preds['Y1'] = train_data['Y1'].mean()
preds['Y2'] = train_data['Y2'].mean()
preds
```

```python
# save preds to csv
preds.to_csv('preds.csv', index=False)
```

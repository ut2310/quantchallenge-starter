```python
from sklearn.linear_model import LinearRegression, RidgeCV, LassoCV, ElasticNetCV
from sklearn.model_selection import TimeSeriesSplit, cross_val_score
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
train_data['O'].head()
train_data['O']  =train_data['O'].fillna(train_data['O'].mean())
```

```python
plt.plot(train_data['O'])
```

```python
train_data.head()
```

```python
import numpy as np
scaler = StandardScaler()
ridge = Pipeline([("scale", scaler), ("ridge", RidgeCV(alphas=np.logspace(-3, 3, 20)))])
lasso = Pipeline([("scale", scaler), ("lasso", LassoCV(alphas=np.logspace(-3, 3, 20), cv=5))])
elastic = Pipeline([
    ("scale", scaler),
    ("elastic", ElasticNetCV(
        alphas=np.logspace(-3, 3, 20),
        l1_ratio=[0.7, 0.8, 0.9, 0.95, 0.99, 1.0],  # search over mixes (0<l1_ratio<=1)
        cv=TimeSeriesSplit(n_splits=5),
        n_jobs=-1,
        max_iter=5000
    ))
])
pca_regression = Pipeline([("scale", scaler), ("PCA", PCA(n_components=0.95)), ("regression", LinearRegression())])
tscv = TimeSeriesSplit(n_splits=5)
models = {
    "Lasso": lasso,
    "Ridge": ridge, 
    "PCA + Regression": pca_regression,
    "Elastic": elastic
}
features = ['G', 'J', 'H', 'C', 'M', 'E', 'N', 'O_logged']
features = features
print(len(features))
X = train_data[features].iloc[:64000]
y = train_data["Y1"].iloc[:64000]
results = {}
for name, model in models.items():
    scores = cross_val_score(model, X, y, cv=tscv, scoring="r2")
    results[name] = scores
    print(f"{name} had r2 score mean {scores.mean()} and score std {scores.std()}")
```

```python
lasso.fit(train_data[features].iloc[:64000], train_data["Y1"].iloc[:64000])
```

```python
y_preds = lasso.predict(train_data[features].iloc[64000:])
plt.plot(train_data['time'].iloc[64000:], y_preds, color='red', linewidth=1.5)
plt.plot(train_data['time'].iloc[64000:], train_data["Y1"].iloc[64000:], color='teal', alpha=0.5, linewidth=1.5)
plt.xlim(64000, 80000)
plt.ylim(-1, 11)
plt.title("Predicted y1 vs Actual Data for Last 16000 timesteps")
```

```python
from sklearn.metrics import r2_score

y_true = train_data["Y1"].iloc[64000:].to_numpy()
r2 = r2_score(y_true, y_preds)
print(f"R^2 = {r2:.6f}")
#with o_logged feature
```

```python
import numpy as np

# from your pipeline:
lasso_cv = lasso.named_steps["lasso"]      # LassoCV instance
sc       = lasso.named_steps["scale"]      # StandardScaler

print("=== LassoCV ===")
print("alpha_*        :", lasso_cv.alpha_)                 # best alpha
print("n_alphas       :", len(lasso_cv.alphas_))           # number tried
print("alphas (first5):", np.round(lasso_cv.alphas_[:5], 6))
print("coef (std.)    :", lasso_cv.coef_)                  # coefficients in scaled space
print("intercept      :", lasso_cv.intercept_)
print("# nonzero coef :", np.count_nonzero(lasso_cv.coef_))

# Optional: quick CV MSE summary (available in LassoCV)
mean_mse = lasso_cv.mse_path_.mean(axis=1)                 # avg over folds
best_i   = int(np.argmin(mean_mse))
print("min mean MSE   :", mean_mse[best_i])
print("alpha@min MSE  :", lasso_cv.alphas_[best_i])
```
